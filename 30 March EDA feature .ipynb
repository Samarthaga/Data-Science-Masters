```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
## Check for multicolinearity 
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: ## Abs = Absolute it makes negative percentage to positive as well 
                colname=corr_matrix.columns[i]
                col_corr.add(colname)   
    return col_corr
```

```python
## Threshold expertise
corr_features=correlation(X_train,0.85)
```

```python
corr_features
```

```python
## BUI and DC are 85 more than 85 per correlated 
```

```python
## drop features when correlation is more than 85 per 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

# Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box Plots To Understand Effects of Standard Scaler 

```python
plt.subplots(figsize=(10,4))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplots(figsize=(10,4))
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
## Box plot on X _train before there is so much diff in respect to datapoints but after scaling the range got in place and data pointslie within same range 
```

# Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linereg=LinearRegression()
linereg.fit(X_train_scaled,y_train)
y_pred=linereg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print('Mean Absolute Error',mae)
print('R2 score',score)
plt.scatter(y_test,y_pred)
```

# Lasso Regression 

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print('Mean Absolute error',mae)
print('R2 Score',score)
plt.scatter(y_test,y_pred)
```
