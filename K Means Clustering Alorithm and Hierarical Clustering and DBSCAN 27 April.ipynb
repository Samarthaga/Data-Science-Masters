# K Means Clustering Algorithms Implementation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 
from sklearn.datasets import make_blobs
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
y[:5] # y is not being used as its unsupervised machine learning and clustering is done for X only 
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow Method to select the K value (Sum of squares will be used)
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## Plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
y_label=kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_label)
```

```python
## Knee Locator 
!pip install kneed
```

```python
from kneed import KneeLocator
```

```python
k1=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
k1.elbow
```

```python
## Silhoutte Score 
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
## Plotting Silsilhouette score
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel('Number of clusters')
plt.ylabel('silhouette_coefficients')
plt.show() # K=3 is the best value 
```

# Hierarical Clustering Implementation 

```python
# Import Iris Dataset
iris=load_iris()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
iris_data.shape
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(iris_data)
```

```python
x_scaled[:6]
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(x_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=iris.target)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
## Agglomerative Clustering
## To construct a dendogram 
import scipy.cluster.hierarchy as sc
## Plot the dendogram 
plt.figure(figsize=(20,7))
plt.title('Dendogram')

# create Dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendogram')
plt.xlabel('Sample Index')
plt.ylabel('Euclian Distance')
plt.xticks(rotation=30)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
## Silhouette Score
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,affinity='euclidean',linkage='ward')
    agglo.fit(x_scaled)
    score=silhouette_score(x_scaled,agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel('Number of clusters')
plt.ylabel('silhouette_coefficients')
plt.show() # K=3 is the best value 
```

# DBscan Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=250,noise=0.15)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_classification=[]
for i in range(-1,2):
    dbscan=DBSCAN(eps=i)
    dbscan.fit(X_scaled)
    score=silhouette_score(X_scaled,dbscan.labels_)
    silhouette_classification.append(score)
```
