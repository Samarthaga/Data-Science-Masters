```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
## now first we find the correlation b/w the data 
plt.figure(figsize=(10,12))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
# it seems there might be colineararity b/w the data 
```

```python
X_train.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X_train,0.85)
```

```python
## seems all these data are corelated and can be dropped from the dataset 
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
## now i will perform the feature scaling for the data 
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train_scaled=scale.fit_transform(X_train,y_train)
X_test_scaled=scale.transform(X_test)
```

```python
## Box plot to understand the data 
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train_scaled,y_train)
```

```python
y_pred=svc.predict(X_test_scaled)
```

```python
svc.coef_
```

```python
from sklearn.metrics import f1_score,accuracy_score,precision_score,recall_score
print(f1_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```
