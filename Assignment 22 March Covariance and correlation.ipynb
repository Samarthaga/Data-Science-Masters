## Covariance and Corelation with Python

```python
import numpy as np
import seaborn as sns 
```

```python
df=sns.load_dataset('healthexp')
```

```python
df.head()
```

```python
## Covariance 
df.cov()
```

```python
## Spearman Rank Correlation
df.corr(method='spearman')
```

```python
## Pearson Rank Correlation
df.corr(method='pearson')
```

```python
## Ans 1: data for students studying for exams and their final results 
import numpy as np
import pandas as pd
df=pd.DataFrame({'hours':[3,4,5,6,7,8],'results':[35,40,50,60,70,80]})
```
