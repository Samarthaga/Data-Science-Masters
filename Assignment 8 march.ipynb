```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

```python
k=np.arange(0,17)
```

```python
k
```

```python
cdf = poisson.cdf(k, mu=7)
cdf = np.round(cdf, 3)

print(cdf)
```

```python
plt.plot(k, cdf, marker='o')
plt.xlabel('k')
plt.ylabel('Cumulative Probability')

plt.show()
```
