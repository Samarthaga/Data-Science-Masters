## 3d Graph

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
axis=fig.add_subplot(projection='3d')
axis.scatter(x,y,z)
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.show()
```

```python
## Ans 4
## Bar plot is used to show relationship b/w numerical and categorical value 
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(3,2))
plt.bar(company,profit,color='r',alpha=0.9)
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.title('company profit bar graph')
plt.grid(c='b',alpha=0.3)
```

```python
plt.figure(figsize=(3,2))
plt.barh(company,profit,color='r',alpha=0.9)
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.title('company profit bar graph')
plt.grid(c='b',alpha=0.3)
```

```python
## Ans 5 
## Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outlier
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(3,2))
plt.boxplot(box1)
plt.show()
plt.figure(figsize=(3,2))
plt.boxplot(box2)
plt.show()
```
