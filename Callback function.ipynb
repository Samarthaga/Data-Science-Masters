```python
# As now we have uploaded the dataset of wine from CSV file , data exploration is done
# Ans Check for null values, identify categorical variables, and encode them.

# first lets explore the dataset 
print(data.info())
# As from the below info all the columns are non null therefore there are no null values , however we should check the min max.

# Null values in the dataset 
data.isnull().sum()
```

```python
# here we can see that there are 2 categorical type data in the dataset 
print(data['quality'].unique())
print('\n ============\n')
# As data balance is very imp. for training and prediction the dataset there are almost equal no of dividition of the dataset 
print(data['quality'].value_counts())
```

```python
data['quality'] = data['quality'].str.upper()
```

```python
data.head(5)
data['quality'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder()
data['quality_1']=encode.fit_transform(data[['quality']]).toarray()
```

```python
print(data.info())
print(data['quality_1'].value_counts())
# now we have encoded the quality feature , we can drop quality categorical feature 
```

```python
data=data.drop(columns=['quality'])
```

```python
# categorical column is dropped and can further be proceeded for train and test split
data.head()
```

```python
# Seperate the features and target variables from the dataframe 
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
# seperation of teh dataframe is done and can be used further for train and test split
print(X.head())
print(y[:2])
```

```python
# Ans 5 : Train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# we can check the shape of train and test data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
sns.boxplot(data)
```

```python
# Ans 6: Scaling on the dataset usuing Standard Scaler
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X_train_scaled=std.fit_transform(X_train)
X_test_scaled=std.transform(X_test)
```

```python
X_train_scaled
sns.boxplot(X_train_scaled)
```

```python
X_train_scaled.shape
```

```python
import keras 
model_clf = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[11]),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Dense(300, activation="elu", kernel_initializer="he_normal"),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Dense(100, activation="elu", kernel_initializer="he_normal"),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Dense(10, activation="softmax")
])
model_clf.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
```

```python
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
```

```python
CKPT_path = "Model_ckpt.h5"
checkpointing_cb = tf.keras.callbacks.ModelCheckpoint(CKPT_path, save_best_only=True)
```

```python
import time 

def get_log_path(log_dir="logs/fit"):
  fileName = time.strftime("log_%Y_%m_%d_%H_%M_%S")
  logs_path = os.path.join(log_dir, fileName)
  print(f"Saving logs at {logs_path}")
  return logs_path

log_dir = get_log_path()
tb_cb = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
```

```python
EPOCHS = 30
VALIDATION_SET = (X_train_scaled, y_train)
```

```python
history = model_clf.fit(X_train_scaled, y_train, epochs=EPOCHS,
                    validation_data=VALIDATION_SET, batch_size=32, callbacks=[tb_cb, early_stopping_cb,checkpointing_cb] )
```
