````python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
````

````python
output_file('test.html')
````

````python
p=figure(title='test flower')
p.xaxis.axis_label ='petal_length'
p.yaxis.axis_label='petal_width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
````

````python
p=figure(title='test flower')
p.xaxis.axis_label ='petal_length'
p.yaxis.axis_label='petal_width'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
````

````python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='linr graph')
p.line(x,y)
show(p)

````

````python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='linr graph')
p.scatter(x,y,size=20,fill_color='red',legend_label='this is my data points')
show(p)
````

````python
fig = figure(plot_width = 200, plot_height = 400)
fig.vbar(x = [1,2,3], width = 0.5, bottom = 0, top = [2,4,6], color = "Cyan")
output_file('bar.html')
show(fig)
````

Ans 3

````python
x1 = [1, 3, 4, 5, 6]
x2 = [5, 3, 8, 1, 8]
y1 = [6, 7, 8, 9, 4]
y2 = [3, 4, 5, 6, 7]
````

````python
p = figure(title="Drawing multiple lines",
           x_axis_label="X-Axis", y_axis_label="Y-Axis")
````

````python
p.line(x1, y1, legend_label="line 1", line_color="red", line_width=1)
p.line(x2, y2, legend_label="line 2", line_color="blue", line_width=1)
p.line(x1, y2, legend_label="line 3", line_color="black", line_width=1)
show(p)
````

Ans 4: Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code.

````python
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import file_html

plot = figure()
plot.circle([1,2], [3,4])

html = file_html(plot, CDN, "my plot")
````
