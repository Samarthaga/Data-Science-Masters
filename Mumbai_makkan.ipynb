```python
data_2=data_1.drop(labels=['BHK'],axis=1)
```

```python
data_2.head()
```

```python
# lets simply change the data type of price_sqft,price,BHK
# first change the price of CR and L to numeric values 
data_2.dtypes
```

```python
# since data is not null we can process the data further 
data_2['price'].isnull().sum()
```

```python
# here we come to know its been written as L and Cr which can be replaced and sliced well 
data_2['price'].value_counts()
```

```python
data_2['price']=data_2['price'].str.replace(' L','00000')
data_2['price']=data_2['price'].str.replace(' Cr','000000')
data_2['price']=data_2['price'].str.split('.').str[1]
data_2['price']=data_2['price'].astype(float)
```

```python
# Price is now float data type
data_2['price'].isnull().sum()
```

```python
# still some of the data is left to work , which is rectified above using the split function where . is replaced to 0000
data_2['price'].value_counts()
```

```python
# we can change the price_sqft , to no space for float data type 
data_2.dtypes
```

```python
# feture type updated is created 
data_2['price_sqft_updated']=data_2['price_sqft'].str.replace(',','')
```
